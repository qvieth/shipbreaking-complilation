```python
## import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data for 1967–2013 with actual Gross Tonnage values
years_1967_2013 = list(range(1967, 2014, 1))  # Every year for data
countries = ['Taiwan', 'Japan', 'USA', 'S. Korea', 'China & Hong Kong', 'India', 'Bangladesh', 'Pakistan']

# Define actual Gross Tonnage values for each country and year (in millions, replace with your data)
gross_tonnage_1967_2013 = {
    1967: {'Taiwan': 0.778, 'Japan': 0.764, 'USA': 0.806, 'S. Korea': 0, 'China & Hong Kong': 0.382, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.145},
    1968: {'Taiwan': 1.118, 'Japan': 0.424, 'USA': 0.636, 'S. Korea': 0, 'China & Hong Kong': 0.339, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.272},
    1969: {'Taiwan': 1.245, 'Japan': 0.424, 'USA': 0.636, 'S. Korea': 0, 'China & Hong Kong': 0.509, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.781},
    1970: {'Taiwan': 1.118, 'Japan': 0.212, 'USA': 0.933, 'S. Korea': 0, 'China & Hong Kong': 0.466, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.697},
    1971: {'Taiwan': 1.372, 'Japan': 0.084, 'USA': 0.763, 'S. Korea': 0, 'China & Hong Kong': 0.296, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.781},
    1972: {'Taiwan': 1.712, 'Japan': 0.127, 'USA': 0.678, 'S. Korea': 0, 'China & Hong Kong': 0.296, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.163},
    1973: {'Taiwan': 1.627, 'Japan': 0.169, 'USA': 0.212, 'S. Korea': 0, 'China & Hong Kong': 0.212, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.357},
    1974: {'Taiwan': 1.245, 'Japan': 0.127, 'USA': 0.212, 'S. Korea': 0, 'China & Hong Kong': 0.339, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.018},
    1975: {'Taiwan': 2.136, 'Japan': 0.127, 'USA': 0.212, 'S. Korea': 0.296, 'China & Hong Kong': 0.169, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.254, 'Others': 1.866},
    1976: {'Taiwan': 3.451, 'Japan': 0.169, 'USA': 0.212, 'S. Korea': 0.424, 'China & Hong Kong': 0.212, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.169, 'Others': 1.994},
    1977: {'Taiwan': 3.366, 'Japan': 0.212, 'USA': 0.127, 'S. Korea': 0.254, 'China & Hong Kong': 0.169, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.254, 'Others': 1.697},
    1978: {'Taiwan': 6.039, 'Japan': 0.359, 'USA': 0.212, 'S. Korea': 0.838, 'China & Hong Kong': 0.339, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.593, 'Others': 1.697},
    1979: {'Taiwan': 4.003, 'Japan': 0.381, 'USA': 0.127, 'S. Korea': 0.424, 'China & Hong Kong': 0.127, 'India': 0.127, 'Bangladesh': 0.0, 'Pakistan': 0.169, 'Others': 1.315},
    1980: {'Taiwan': 4.427, 'Japan': 0.1, 'USA': 0.127, 'S. Korea': 0.169, 'China & Hong Kong': 1.7, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.296, 'Others': 0.806},
    1981: {'Taiwan': 5.275, 'Japan': 0.127, 'USA': 0.254, 'S. Korea': 0.763, 'China & Hong Kong': 0, 'India': 0.212, 'Bangladesh': 0.0, 'Pakistan': 0.763, 'Others': 0.678},
    1982: {'Taiwan': 7.821, 'Japan': 0.296, 'USA': 0, 'S. Korea': 2.3, 'China & Hong Kong': 0.212, 'India': 0.254, 'Bangladesh': 0.0, 'Pakistan': 1.569, 'Others': 1.23},
    1983: {'Taiwan': 7.863, 'Japan': 0.466, 'USA': 0, 'S. Korea': 3.945, 'China & Hong Kong': 0.212, 'India': 0.509, 'Bangladesh': 0.339, 'Pakistan': 1.739, 'Others': 1.739},
    1984: {'Taiwan': 6.718, 'Japan': 0.89, 'USA': 0, 'S. Korea': 4.157, 'China & Hong Kong': 2.078, 'India': 0.509, 'Bangladesh': 0.678, 'Pakistan': 0.975, 'Others': 1.782},
    1985: {'Taiwan': 7.863, 'Japan': 0.933, 'USA': 0, 'S. Korea': 2.63, 'China & Hong Kong': 4.963, 'India': 1.357, 'Bangladesh': 0.806, 'Pakistan': 1.145, 'Others': 2.63},
    1986: {'Taiwan': 7.821, 'Japan': 0.721, 'USA': 0, 'S. Korea': 2.672, 'China & Hong Kong': 4.497, 'India': 0.636, 'Bangladesh': 0.254, 'Pakistan': 0.806, 'Others': 3.012},
    1987: {'Taiwan': 4.469, 'Japan': 0.593, 'USA': 0, 'S. Korea': 1.272, 'China & Hong Kong': 1.824, 'India': 1.697, 'Bangladesh': 0.339, 'Pakistan': 0.636, 'Others': 1.315},
    1988: {'Taiwan': 1.542, 'Japan': 0.127, 'USA': 0.212, 'S. Korea': 0.466, 'China & Hong Kong': 0.509, 'India': 0.466, 'Bangladesh': 0.509, 'Pakistan': 0.212, 'Others': 1.484},
    1989: {'Taiwan': 0.142, 'Japan': 0.254, 'USA': 0, 'S. Korea': 0.424, 'China & Hong Kong': 0.721, 'India': 0.339, 'Bangladesh': 0.123, 'Pakistan': 0.0, 'Others': 0.636},
    1990: {'Taiwan': 0.057, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 0.127, 'India': 1.103, 'Bangladesh': 0.169, 'Pakistan': 0.0, 'Others': 0.424},
    1991: {'Taiwan': 0.142, 'Japan': 0, 'USA': 0, 'S. Korea': 0.169, 'China & Hong Kong': 0.721, 'India': 0.509, 'Bangladesh': 0.0, 'Pakistan': 0.381, 'Others': 0.169},
    1992: {'Taiwan': 0.142, 'Japan': 0, 'USA': 0, 'S. Korea': 1.5, 'China & Hong Kong': 2.121, 'India': 2.036, 'Bangladesh': 1.145, 'Pakistan': 0.678, 'Others': 0.254},
    1993: {'Taiwan': 0.184, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 5.685, 'India': 1.994, 'Bangladesh': 1.400, 'Pakistan': 0.890, 'Others': 0.636},
    1994: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 2.518, 'India': 2.503, 'Bangladesh': 2.078, 'Pakistan': 2.163, 'Others': 0.466},
    1995: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0.269, 'China & Hong Kong': 0.731, 'India': 2.800, 'Bangladesh': 2.545, 'Pakistan': 1.654, 'Others': 0.593},
    1996: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 0.142, 'India': 4.921, 'Bangladesh': 2.163, 'Pakistan': 2.121, 'Others': 0.339},
    1997: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 0.227, 'India': 4.879, 'Bangladesh': 2.291, 'Pakistan': 0.848, 'Others': 0.509},
    1998: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 0.608, 'India': 6.321, 'Bangladesh': 2.588, 'Pakistan': 2.206, 'Others': 0.636},
    1999: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 2.630, 'India': 6.788, 'Bangladesh': 3.733, 'Pakistan': 2.333, 'Others': 0.721},
    2000: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 2.730, 'India': 5.854, 'Bangladesh': 2.503, 'Pakistan': 0.806, 'Others': 1.357},
    2001: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 2.625, 'India': 4.751, 'Bangladesh': 5.006, 'Pakistan': 1.697, 'Others': 0.424},
    2002: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 3.154, 'India': 6.830, 'Bangladesh': 4.921, 'Pakistan': 0.890, 'Others': 0.636},
    2003: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 5.657, 'India': 5.897, 'Bangladesh': 2.885, 'Pakistan': 0.806, 'Others': 0.721},
    2004: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 1.584, 'India': 1.612, 'Bangladesh': 3.351, 'Pakistan': 0.169, 'Others': 0.551},
    2005: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 0.227, 'India': 1.103, 'Bangladesh': 2.036, 'Pakistan': 0.1, 'Others': 0.424},
    2006: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 0.227, 'India': 0.890, 'Bangladesh': 2.842, 'Pakistan': 0.212, 'Others': 0.466},
    2007: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 0.354, 'India': 1.357, 'Bangladesh': 1.824, 'Pakistan': 0.296, 'Others': 0.381},
    2008: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 1.033, 'India': 2.418, 'Bangladesh': 4.115, 'Pakistan': 0.254, 'Others': 0.551},
    2009: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 7.483, 'India': 7.655, 'Bangladesh': 6.451, 'Pakistan': 2.236, 'Others': 1.118},
    2010: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 4.731, 'India': 6.537, 'Bangladesh': 3.956, 'Pakistan': 2.494, 'Others': 0.946},
    2011: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 5.763, 'India': 8.602, 'Bangladesh': 5.763, 'Pakistan': 3.182, 'Others': 1.634},
    2012: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 8.086, 'India': 12.129, 'Bangladesh': 8.774, 'Pakistan': 5.591, 'Others': 2.064},
    2013: {'Taiwan': 0, 'Japan': 0, 'USA': 0, 'S. Korea': 0, 'China & Hong Kong': 6.881, 'India': 8.430, 'Bangladesh': 7.225, 'Pakistan': 5.161, 'Others': 1.720}
}

# Create DataFrame for 1967–2013 with actual Gross Tonnage values
df_1967_2013 = pd.DataFrame(gross_tonnage_1967_2013).T
df_1967_2013.index = df_1967_2013.index.astype(int)  # Ensure index is integer for consistency

# Verify the DataFrame
print(df_1967_2013.head())  # Check the first few rows to ensure data is correct

# Data for 2014–2023 (from shipbreaking.csv)
df_2014_2023 = pd.read_csv('AA_shipbreaking.csv')
df_2014_2023 = df_2014_2023[df_2014_2023['Year'].between(2014, 2023)]

# Pivot table for 2014–2023: Years as rows, Economy Labels as columns, GT as values (in millions)
gt_df_2014_2023 = df_2014_2023.pivot(index='Year', columns='Economy Label', values='Gross Tonnage') / 1000000

# Map the economies to the country names used in the first chart
country_mapping = {
    'China, Taiwan Province of': 'Taiwan',
    'Japan': 'Japan',
    'United States': 'USA',
    'Republic of Korea': 'S. Korea',
    'China': 'China & Hong Kong',  # Assuming China includes Hong Kong
    'India': 'India',
    'Bangladesh': 'Bangladesh',
    'Pakistan': 'Pakistan'
}

# Extract GT for relevant countries, filling NaN with 0
country_gt_2014_2023 = gt_df_2014_2023[list(country_mapping.keys())].fillna(0)
country_gt_2014_2023.columns = [country_mapping[col] for col in country_gt_2014_2023.columns]

# Calculate "Others" as World total minus sum of selected countries for 2014–2023
world_gt_2014_2023 = gt_df_2014_2023['World']
others_gt_2014_2023 = world_gt_2014_2023 - country_gt_2014_2023.sum(axis=1)
country_gt_2014_2023['Others'] = others_gt_2014_2023

# Combine the DataFrames (1967–2013 and 2014–2023), filling the gap (2013–2013) and 2024 with zeros
all_years = list(range(1967, 2025))  # Every year from 1967 to 2024
combined_df = pd.DataFrame(index=all_years, columns=countries + ['Others'])

# Fill with 1967–2013 data
combined_df.loc[years_1967_2013] = df_1967_2013

# Fill with 2014–2023 data (mapped to the country names)
combined_df.loc[country_gt_2014_2023.index] = country_gt_2014_2023

# Fill the gap (2013–2013) and 2024 with zeros
gap_years = [y for y in all_years if y not in combined_df.index]
combined_df.loc[gap_years] = 0

# Define colors and patterns from the first chart
color_dict = {
    'Taiwan': 'lightblue',
    'Japan': 'yellow',
    'USA': 'lightyellow',
    'S. Korea': 'darkblue',
    'China & Hong Kong': 'red',
    'India': 'blue',
    'Bangladesh': 'purple',
    'Pakistan': 'lightgreen',
    'Others': 'black'
}

hatch_dict = {
    'Taiwan': '/',
    'Japan': None,
    'USA': None,
    'S. Korea': None,
    'China & Hong Kong': None,
    'India': '-',
    'Bangladesh': '|',
    'Pakistan': None,
    'Others': None
}

# Plotting
fig, ax = plt.subplots(figsize=(15, 6))  # Increased figure width for more space
bar_width = 0.8
years = combined_df.index
bottom = np.zeros(len(years))

# Plot each country's contribution as a stacked segment
for country in countries + ['Others']:
    ax.bar(
        years,
        combined_df[country],
        bottom=bottom,
        label=country,
        color=color_dict[country],
        hatch=hatch_dict[country],
        width=bar_width
    )
    bottom += combined_df[country]

# Customize the plot (matching the first chart's style, with x-axis changes)
ax.set_xlabel('Year')
ax.set_ylabel('Million Gross Tonnages')
ax.set_title('Annual Shipbreaking Output in Million Gross Tonnages (1967-2023)')
ax.set_yticks(range(0, 41, 5))  # Match the original y-axis range (0–25, step 5)
ax.set_xticks(range(1967, 2025, 4))  # Show ticks every 4 years for more space between labels
ax.set_xticklabels(range(1967, 2025, 4), rotation=0)  # Rotate 0 degrees (horizontal)
ax.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
ax.legend(loc='lower left', bbox_to_anchor=(0, -0.3), ncol=3)

# Add source note below the chart
# plt.figtext(0.1, -0.2, 
#           "Source: Based on data from Shipbuilder’s Association of Japan (eds.) Shipbuilding Statistics 2009, "
#           "Lloyd’s Register, Casualty Return, various years, and shipbreaking.csv.\n"
#           "Note: Covers commercial ships larger than 100 Gross Tonnages. "
#           "Gross Tonnage (GT) is a unit based on the volume of the vessel.", 
#           wrap=True, horizontalalignment='left', fontsize=8)

plt.tight_layout()

plt.savefig('shipbuilding_output.png', bbox_inches='tight')

plt.show()

```
