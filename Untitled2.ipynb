```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read data from the CSV file in the same folder
df = pd.read_csv('AA_shipbreaking.csv')

# Filter for 2014-2023
df = df[df['Year'].between(2014, 2023)]

# Pivot table: Years as rows, Economy Labels as columns, GT as values (in millions)
gt_df = df.pivot(index='Year', columns='Economy Label', values='Gross Tonnage') / 1000000

# Define countries and extract their GT, filling NaN with 0
countries = [
    'China, Taiwan Province of',  # Taiwan
    'Japan',
    'United States',             # USA
    'Republic of Korea',         # South Korea
    'China',                     # China & Hong Kong
    'India',
    'Bangladesh',
    'Pakistan'
]
country_gt = gt_df[countries].fillna(0)

# Calculate "Others" as World total minus sum of selected countries
world_gt = gt_df['World']
others_gt = world_gt - country_gt.sum(axis=1)

# Prepare plotting DataFrame
plot_df = country_gt.copy()
plot_df['Others'] = others_gt

# Define stacking order (bottom to top) as per the first code
stack_order = [
    'China, Taiwan Province of',  # Taiwan
    'Japan',
    'United States',             # USA
    'Republic of Korea',         # South Korea
    'China',                     # China & Hong Kong
    'India',
    'Bangladesh',
    'Pakistan',
    'Others'
]

# Define colors and patterns from the first code
color_dict = {
    'China, Taiwan Province of': 'lightblue',  # Taiwan
    'Japan': 'yellow',
    'United States': 'lightyellow',           # USA
    'Republic of Korea': 'darkblue',          # South Korea
    'China': 'red',                           # China & Hong Kong
    'India': 'blue',
    'Bangladesh': 'purple',
    'Pakistan': 'lightgreen',
    'Others': 'black'
}

hatch_dict = {
    'China, Taiwan Province of': '/',
    'Japan': None,
    'United States': None,
    'Republic of Korea': None,
    'China': None,
    'India': '-',
    'Bangladesh': '|',
    'Pakistan': None,
    'Others': None
}

# Plotting
fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.8  # Match the width from the first code
years = plot_df.index
bottom = np.zeros(len(years))

for country in stack_order:
    ax.bar(
        years,
        plot_df[country],
        bottom=bottom,
        label=country,
        color=color_dict[country],
        hatch=hatch_dict[country],
        width=bar_width
    )
    bottom += plot_df[country]

# Customize the plot (matching the first code's style)
ax.set_xlabel('Year')
ax.set_ylabel('Million Gross Tonnages')
ax.set_title('Annual Shipbuilding Output in Million Gross Tonnages (2014-2023)')
ax.set_yticks(range(0, 26, 5))  # Match the original y-axis range (025, step 5)
ax.set_xticks(years)
ax.set_xticklabels(years, rotation=90)  # Rotate 90 degrees as in the first code
ax.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
ax.legend(loc='lower left', bbox_to_anchor=(0, -0.3), ncol=3)

# Add source note below the chart (adjust as needed based on your data source)
plt.figtext(0.1, -0.2, 
            "Source: Based on data from shipbreaking.csv.\n"
            "Note: Covers commercial ships larger than 100 Gross Tonnages. "
            "Gross Tonnage (GT) is a unit based on the volume of the vessel.", 
            wrap=True, horizontalalignment='left', fontsize=8)

plt.tight_layout()
plt.show()
```
