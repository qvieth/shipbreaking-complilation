```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data for 1967–2007 with actual Gross Tonnage values
years_1967_2007 = list(range(1967, 2008, 1))  # Every year for data
countries = ['Taiwan', 'Japan', 'USA', 'S. Korea', 'China & Hong Kong', 'India', 'Bangladesh', 'Pakistan']

# Define actual Gross Tonnage values for each country and year (in millions, replace with your data)
gross_tonnage_1967_2007 = {
    1967: {'Taiwan': 2.5, 'Japan': 2.0, 'USA': 0.5, 'S. Korea': 0.25, 'China & Hong Kong': 0.25, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.5},
    1968: {'Taiwan': 0.5, 'Japan': 2.0, 'USA': 0.5, 'S. Korea': 0.25, 'China & Hong Kong': 0.25, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.5},
    1969: {'Taiwan': 0.5, 'Japan': 2.0, 'USA': 0.5, 'S. Korea': 0.25, 'China & Hong Kong': 0.25, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.5},
    1970: {'Taiwan': 0.5, 'Japan': 2.0, 'USA': 0.5, 'S. Korea': 0.25, 'China & Hong Kong': 0.25, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.5},
    1971: {'Taiwan': 0.5, 'Japan': 2.0, 'USA': 0.5, 'S. Korea': 0.25, 'China & Hong Kong': 0.25, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.5},
    1972: {'Taiwan': 0.5, 'Japan': 2.0, 'USA': 0.5, 'S. Korea': 0.25, 'China & Hong Kong': 0.25, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.5},
    1973: {'Taiwan': 0.5, 'Japan': 2.0, 'USA': 0.5, 'S. Korea': 0.25, 'China & Hong Kong': 0.25, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.5},
    1974: {'Taiwan': 0.7, 'Japan': 2.8, 'USA': 0.7, 'S. Korea': 0.35, 'China & Hong Kong': 0.35, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.1},
    1975: {'Taiwan': 0.5, 'Japan': 3.5, 'USA': 0.5, 'S. Korea': 3.0, 'China & Hong Kong': 1.0, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.0},
    1976: {'Taiwan': 0.55, 'Japan': 3.85, 'USA': 0.55, 'S. Korea': 3.3, 'China & Hong Kong': 1.1, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.2},
    1977: {'Taiwan': 0.65, 'Japan': 4.55, 'USA': 0.65, 'S. Korea': 3.9, 'China & Hong Kong': 1.3, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.6},
    1978: {'Taiwan': 0.7, 'Japan': 4.9, 'USA': 0.7, 'S. Korea': 4.2, 'China & Hong Kong': 1.4, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.8},
    1979: {'Taiwan': 0.8, 'Japan': 5.6, 'USA': 0.8, 'S. Korea': 4.8, 'China & Hong Kong': 1.6, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 3.2},
    1980: {'Taiwan': 0.85, 'Japan': 5.95, 'USA': 0.85, 'S. Korea': 5.1, 'China & Hong Kong': 1.7, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 3.4},
    1981: {'Taiwan': 0.95, 'Japan': 6.65, 'USA': 0.95, 'S. Korea': 5.7, 'China & Hong Kong': 1.9, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 3.8},
    1982: {'Taiwan': 1.0, 'Japan': 7.0, 'USA': 1.0, 'S. Korea': 6.0, 'China & Hong Kong': 2.0, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 4.0},
    1983: {'Taiwan': 1.1, 'Japan': 7.7, 'USA': 1.1, 'S. Korea': 6.6, 'China & Hong Kong': 2.2, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 4.4},
    1984: {'Taiwan': 1.15, 'Japan': 8.05, 'USA': 1.15, 'S. Korea': 6.9, 'China & Hong Kong': 2.3, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 4.6},
    1985: {'Taiwan': 1.25, 'Japan': 8.75, 'USA': 1.25, 'S. Korea': 7.5, 'China & Hong Kong': 2.5, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 5.0},
    1986: {'Taiwan': 1.1, 'Japan': 7.7, 'USA': 1.1, 'S. Korea': 6.6, 'China & Hong Kong': 2.2, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 4.4},
    1987: {'Taiwan': 0.75, 'Japan': 5.25, 'USA': 0.75, 'S. Korea': 4.5, 'China & Hong Kong': 1.5, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 3.0},
    1988: {'Taiwan': 0.6, 'Japan': 4.2, 'USA': 0.6, 'S. Korea': 3.6, 'China & Hong Kong': 1.2, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.4},
    1989: {'Taiwan': 0.5, 'Japan': 3.5, 'USA': 0.5, 'S. Korea': 3.0, 'China & Hong Kong': 1.0, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 2.0},
    1990: {'Taiwan': 0.35, 'Japan': 2.45, 'USA': 0.35, 'S. Korea': 2.1, 'China & Hong Kong': 0.7, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.4},
    1991: {'Taiwan': 0.25, 'Japan': 1.75, 'USA': 0.25, 'S. Korea': 1.5, 'China & Hong Kong': 0.5, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.0},
    1992: {'Taiwan': 0.25, 'Japan': 1.75, 'USA': 0.25, 'S. Korea': 1.5, 'China & Hong Kong': 0.5, 'India': 0.0, 'Bangladesh': 0.0, 'Pakistan': 0.0, 'Others': 1.0},
    1993: {'Taiwan': 0.25, 'Japan': 1.0, 'USA': 0.1, 'S. Korea': 2.0, 'China & Hong Kong': 1.0, 'India': 0.1, 'Bangladesh': 0.1, 'Pakistan': 0.05, 'Others': 0.4},
    1994: {'Taiwan': 0.3, 'Japan': 1.2, 'USA': 0.12, 'S. Korea': 2.4, 'China & Hong Kong': 1.2, 'India': 0.12, 'Bangladesh': 0.12, 'Pakistan': 0.06, 'Others': 0.48},
    1995: {'Taiwan': 0.35, 'Japan': 1.4, 'USA': 0.14, 'S. Korea': 2.8, 'China & Hong Kong': 1.4, 'India': 0.14, 'Bangladesh': 0.14, 'Pakistan': 0.07, 'Others': 0.56},
    1996: {'Taiwan': 0.4, 'Japan': 1.6, 'USA': 0.16, 'S. Korea': 3.2, 'China & Hong Kong': 1.6, 'India': 0.16, 'Bangladesh': 0.16, 'Pakistan': 0.08, 'Others': 0.64},
    1997: {'Taiwan': 0.5, 'Japan': 2.0, 'USA': 0.2, 'S. Korea': 4.0, 'China & Hong Kong': 2.0, 'India': 0.2, 'Bangladesh': 0.2, 'Pakistan': 0.1, 'Others': 0.8},
    1998: {'Taiwan': 0.6, 'Japan': 2.4, 'USA': 0.24, 'S. Korea': 4.8, 'China & Hong Kong': 2.4, 'India': 0.24, 'Bangladesh': 0.24, 'Pakistan': 0.12, 'Others': 0.96},
    1999: {'Taiwan': 0.75, 'Japan': 3.0, 'USA': 0.3, 'S. Korea': 6.0, 'China & Hong Kong': 3.0, 'India': 0.3, 'Bangladesh': 0.3, 'Pakistan': 0.15, 'Others': 1.2},
    2000: {'Taiwan': 0.85, 'Japan': 3.4, 'USA': 0.34, 'S. Korea': 6.8, 'China & Hong Kong': 3.4, 'India': 0.34, 'Bangladesh': 0.34, 'Pakistan': 0.17, 'Others': 1.36},
    2001: {'Taiwan': 1.0, 'Japan': 4.0, 'USA': 0.4, 'S. Korea': 8.0, 'China & Hong Kong': 4.0, 'India': 0.4, 'Bangladesh': 0.4, 'Pakistan': 0.2, 'Others': 1.6},
    2002: {'Taiwan': 0.95, 'Japan': 3.8, 'USA': 0.38, 'S. Korea': 7.6, 'China & Hong Kong': 3.8, 'India': 0.38, 'Bangladesh': 0.38, 'Pakistan': 0.19, 'Others': 1.52},
    2003: {'Taiwan': 0.9, 'Japan': 3.6, 'USA': 0.36, 'S. Korea': 7.2, 'China & Hong Kong': 3.6, 'India': 0.36, 'Bangladesh': 0.36, 'Pakistan': 0.18, 'Others': 1.44},
    2004: {'Taiwan': 0.85, 'Japan': 3.4, 'USA': 0.34, 'S. Korea': 6.8, 'China & Hong Kong': 3.4, 'India': 0.34, 'Bangladesh': 0.34, 'Pakistan': 0.17, 'Others': 1.36},
    2005: {'Taiwan': 0.8, 'Japan': 3.2, 'USA': 0.32, 'S. Korea': 6.4, 'China & Hong Kong': 3.2, 'India': 0.32, 'Bangladesh': 0.32, 'Pakistan': 0.16, 'Others': 1.28},
    2006: {'Taiwan': 0.75, 'Japan': 3.0, 'USA': 0.3, 'S. Korea': 6.0, 'China & Hong Kong': 3.0, 'India': 0.3, 'Bangladesh': 0.3, 'Pakistan': 0.15, 'Others': 1.2},
    2007: {'Taiwan': 0.75, 'Japan': 3.0, 'USA': 0.3, 'S. Korea': 6.0, 'China & Hong Kong': 3.0, 'India': 0.3, 'Bangladesh': 0.3, 'Pakistan': 0.15, 'Others': 1.2}
}

# Create DataFrame for 1967–2007 with actual Gross Tonnage values
df_1967_2007 = pd.DataFrame(gross_tonnage_1967_2007).T
df_1967_2007.index = df_1967_2007.index.astype(int)  # Ensure index is integer for consistency

# Verify the DataFrame
print(df_1967_2007.head())  # Check the first few rows to ensure data is correct

# Data for 2014–2023 (from shipbreaking.csv)
df_2014_2023 = pd.read_csv('AA_shipbreaking.csv')
df_2014_2023 = df_2014_2023[df_2014_2023['Year'].between(2014, 2023)]

# Pivot table for 2014–2023: Years as rows, Economy Labels as columns, GT as values (in millions)
gt_df_2014_2023 = df_2014_2023.pivot(index='Year', columns='Economy Label', values='Gross Tonnage') / 1000000

# Map the economies to the country names used in the first chart
country_mapping = {
    'China, Taiwan Province of': 'Taiwan',
    'Japan': 'Japan',
    'United States': 'USA',
    'Republic of Korea': 'S. Korea',
    'China': 'China & Hong Kong',  # Assuming China includes Hong Kong
    'India': 'India',
    'Bangladesh': 'Bangladesh',
    'Pakistan': 'Pakistan'
}

# Extract GT for relevant countries, filling NaN with 0
country_gt_2014_2023 = gt_df_2014_2023[list(country_mapping.keys())].fillna(0)
country_gt_2014_2023.columns = [country_mapping[col] for col in country_gt_2014_2023.columns]

# Calculate "Others" as World total minus sum of selected countries for 2014–2023
world_gt_2014_2023 = gt_df_2014_2023['World']
others_gt_2014_2023 = world_gt_2014_2023 - country_gt_2014_2023.sum(axis=1)
country_gt_2014_2023['Others'] = others_gt_2014_2023

# Combine the DataFrames (1967–2007 and 2014–2023), filling the gap (2008–2013) and 2024 with zeros
all_years = list(range(1967, 2025))  # Every year from 1967 to 2024
combined_df = pd.DataFrame(index=all_years, columns=countries + ['Others'])

# Fill with 1967–2007 data
combined_df.loc[years_1967_2007] = df_1967_2007

# Fill with 2014–2023 data (mapped to the country names)
combined_df.loc[country_gt_2014_2023.index] = country_gt_2014_2023

# Fill the gap (2008–2013) and 2024 with zeros
gap_years = [y for y in all_years if y not in combined_df.index]
combined_df.loc[gap_years] = 0

# Define colors and patterns from the first chart
color_dict = {
    'Taiwan': 'lightblue',
    'Japan': 'yellow',
    'USA': 'lightyellow',
    'S. Korea': 'darkblue',
    'China & Hong Kong': 'red',
    'India': 'blue',
    'Bangladesh': 'purple',
    'Pakistan': 'lightgreen',
    'Others': 'black'
}

hatch_dict = {
    'Taiwan': '/',
    'Japan': None,
    'USA': None,
    'S. Korea': None,
    'China & Hong Kong': None,
    'India': '-',
    'Bangladesh': '|',
    'Pakistan': None,
    'Others': None
}

# Plotting
fig, ax = plt.subplots(figsize=(15, 6))  # Increased figure width for more space
bar_width = 0.8
years = combined_df.index
bottom = np.zeros(len(years))

# Plot each country's contribution as a stacked segment
for country in countries + ['Others']:
    ax.bar(
        years,
        combined_df[country],
        bottom=bottom,
        label=country,
        color=color_dict[country],
        hatch=hatch_dict[country],
        width=bar_width
    )
    bottom += combined_df[country]

# Customize the plot (matching the first chart's style, with x-axis changes)
ax.set_xlabel('Year')
ax.set_ylabel('Million Gross Tonnages')
ax.set_title('Annual Shipbuilding Output in Million Gross Tonnages (1967-2024)')
ax.set_yticks(range(0, 26, 5))  # Match the original y-axis range (0–25, step 5)
ax.set_xticks(range(1967, 2025, 4))  # Show ticks every 4 years for more space between labels
ax.set_xticklabels(range(1967, 2025, 4), rotation=0)  # Rotate 0 degrees (horizontal)
ax.grid(axis='y', color='gray', linestyle='--', linewidth=0.5)
ax.legend(loc='lower left', bbox_to_anchor=(0, -0.3), ncol=3)

# Add source note below the chart
# plt.figtext(0.1, -0.2, 
#           "Source: Based on data from Shipbuilder’s Association of Japan (eds.) Shipbuilding Statistics 2009, "
#           "Lloyd’s Register, Casualty Return, various years, and shipbreaking.csv.\n"
#           "Note: Covers commercial ships larger than 100 Gross Tonnages. "
#           "Gross Tonnage (GT) is a unit based on the volume of the vessel.", 
#           wrap=True, horizontalalignment='left', fontsize=8)

plt.tight_layout()

plt.savefig('shipbuilding_output.png', bbox_inches='tight')

plt.show()

```
